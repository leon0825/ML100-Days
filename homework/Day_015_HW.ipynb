```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = -1 + 2 * np.random.rand(10, 10)

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix, vmin = -1, vmax = 1, cmap = plt.cm.RdYlBu_r, annot = True)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = -1 + 2 * np.random.rand(nrow, ncol)

indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)


grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
'''
grid.map_upper(... , alpha = 0.2)
grid.map_diag(... )
grid.map_lower(... , cmap = plt.cm.OrRd_r)
'''
# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)
# 對角線畫 hist
grid.map_diag(plt.hist)
# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

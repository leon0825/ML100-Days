```python
# info 可以看有幾筆資料、每個欄位的資料型別是什麼(int, float..)、有無空值(null)的存在、佔據多少記憶體
app_train.info()

# 利用 describe 來看資料的平均值、分佈情況、是否有資料傾斜Skew的問題
app_train.describe()
```

```python
column = 'AMT_INCOME_TOTAL'

print('{0} 平均數: {1:.0f}'.format(column, app_train[column].mean()))
print('{0} 標準差: {1:.0f}'.format(column, app_train[column].std()))
```

```python
column = 'AMT_CREDIT'

app_train[column].hist(bins=50)
xmin = app_train[column].min()
xmax = app_train[column].max()
plt.xlim(xmin, xmax)
plt.show()
```
